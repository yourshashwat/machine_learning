```python
df = df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df.iloc[:, 0:2], df.iloc[:, -1], train_size=0.8)
```

```python
#X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
X_train
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
X_train
```

```python
X_train = oe.transform(X_train)
X_test= oe.transform(X_test)
```

```python
oe.categories_
```

```python
X_train
```

```python
X_test
```

```python
le = LabelEncoder()
```

```python
le.fit(Y_train)
```
